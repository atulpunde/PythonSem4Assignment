```python
import requests, time, random   #importing required libraries
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
driver = webdriver.Firefox( executable_path = "E://geckodriver.exe")  #open firefox new tab
link = 'https://collegedunia.com/college/1284-fergusson-college-pune'    #search any link
driver.get(link)     # This require some time to open link.. Please see progressbar..
```

```python
src = driver.page_source      #inspect webpage and take page sourse using driver to extract soup
soup = BeautifulSoup(src, 'lxml')     #using BeautifulSoup we get all soup/data
soup
```

```python
name_div = soup.find('div', {'class': 'content_p'})     #usign soup find data
name_div
```

```python
para1 = name_div.find('p').get_text().strip()      #We can extract whole paragraph
para1
```

```python
table = soup.find('table',{'class':'table table-striped style_table table-nowrap'})   #find tables
table
```

```python
heading = table.findAll('th')       #find column names
# print(heading)
for i in heading:
    print(i.get_text().strip())
```

```python
data = table.findAll('td')       #find table data.. rows and column
# print(data)
count = 0
for i in data:
    print(i.get_text().strip(),end="\t")
    
    if count==3:
        count = 0
        print("\n")
    else:
        count=count+1
```
